__3. Create physical performance reports for the home team players.__

A player's physical performance report would include information such as minutes played, total distance covered, distance covered in different activities (like walking, jogging, running, sprinting), and total number of sprints made. Let's begin by creating a new dataframe to store all of this information for the home team players.

```python
home_players = np.unique([ c.split('_')[1] for c in tracking_home.columns if c[:4] == 'Home' ]) # finding and storing the home team players' kit numbers from their tracking data
home_summary = pd.DataFrame(index=home_players) # creating a new dataframe with the home team kit numbers as the index entries
home_summary.index.name = 'Kit Number' # setting the index name for this new dataframe
```

```python
home_summary
```

Now, we will calculate the minutes played by each home team player and then add this data to the home_summary dataframe.

For the calculation part, here's our approach:

* For each home team player, we will find the first and the last frame in which we have non-null position observations for that player (when a player is not on the pitch positions are null). We can use either the x-coordinate values or the y-coordinate values for this. We will then use this information to calculate the total number of frames each home team player was on the field. 


* Since 25 frames are captured per second, each frame corresponds to 1/25 second. A minute consists of 60 seconds, therefore, each second is equal to 1/60 minute. Using this information, we can calculate the total number of minutes played by a player by multiplying the number of frames he was on the field with ((1/25) * (1/60)). This will give us the minutes played by that player.

Let's do this using Python.

```python
minutes = []
for player in home_players:
    column = 'Home_' + player + '_x' # using the x-coordinate value of the player's position
    player_minutes = (tracking_home[column].last_valid_index() - tracking_home[column].first_valid_index() + 1) / 25 / 60.
    minutes.append(player_minutes)
home_summary['Minutes Played'] = minutes # adding 'Minutes Played' column to the home_summary dataframe
home_summary.sort_values(['Minutes Played'], ascending=False, inplace=True) # sorting the home team players by minutes played in decreasing order
```

```python
home_summary
```

Here, we can observe that 8 players had played the complete match. 3 players were substituted. These substitutions were most probably: #10 out #14 in, #6 out #13 in, and #12 out #1 in.

Next up, we will calculate the total distance (in kilometres) covered by each home team player. Here's our approach for the calculation part:

* Speed is equal to distance/time. While calculating frame-wise player velocities, we used the distance covered by the player (separately in x and y directions) and the total time it took between 2 frames. For calculating the total distance covered by the player (in km), we will multiply the frame-wise speeds with the time period between 2 frames (it's a constant equal to 1/25 second) and divide the product by 1000 to convert the value from metres to kilometres (1 km = 1,000 m). Then, we will take the sum of all of these frame-wise distance values to get the total distance covered by the player.

Let's do this in code.

```python
distance = []
for player in home_summary.index: # using home_summary.index instead of home_players because the index order was changed when we calculated minutes played
    column = 'Home_' + player + '_speed'
    player_distance = (tracking_home[column]/25./1000).sum()
    distance.append(player_distance) 
home_summary['Distance [km]'] = distance
```

```python
home_summary
```

Let's make a bar plot for the total distance covered by each of the home team players.

```python
sns.barplot(x=home_summary.index, y=home_summary['Distance [km]'], color='blue')
plt.title('Distance Covered [km] by Home Team Players')
plt.xlabel('Player Kit Number')
plt.ylabel('Distance Covered [km]')
plt.show()
```

Footballers don't cover these distances with the same pace and the same intensity all the time. Their physical activities, when it comes to covering distances, include walking, jogging, running and sprinting. So you see, they are not running all the time. For the home team players, let's find out how much of these distances are covered while walking, jogging, running and sprinting. First of all, we will set some conditions for this classification:

* If __player's speed < 2 m/s__, then he's __walking__.
* If __2 m/s <= player's speed < 4 m/s__, then he's __jogging__.
* If __4 m/s <= player's speed < 7 m/s__, then he's __running__.
* If __7 m/s < player's speed__, then he's __sprinting__.

The approach to calculate these distances is exactly how we calculated the total distances covered by the players.

```python
walking = []
jogging = []
running = []
sprinting = []
for player in home_summary.index:
    column = 'Home_' + player + '_speed'
    player_distance = (tracking_home[tracking_home[column] < 2][column]/25./1000).sum()
    walking.append(player_distance)
    player_distance = (tracking_home[(tracking_home[column] >= 2) & (tracking_home[column] < 4)][column]/25./1000).sum()
    jogging.append(player_distance)
    player_distance = (tracking_home[(tracking_home[column] >= 4) & (tracking_home[column] < 7)][column]/25./1000).sum()
    running.append(player_distance)
    player_distance = (tracking_home[tracking_home[column] >= 7][column]/25./1000).sum()
    sprinting.append(player_distance)
home_summary['Walking [km]'] = walking
home_summary['Jogging [km]'] = jogging
home_summary['Running [km]'] = running
home_summary['Sprinting [km]'] = sprinting
```

Let's create a clustered bar plot for the distances covered by each of the home team players while walking, jogging, running and sprinting.

```python
home_summary[['Walking [km]','Jogging [km]','Running [km]','Sprinting [km]']].plot.bar(colormap='viridis_r', figsize=(12, 8))
plt.title('Distance Covered [km] by Home Team Players while Walking, Jogging, Running and Sprinting')
plt.xlabel('Player Kit Number')
plt.ylabel('Distance Covered [km]')
plt.show()
```

Finally, we will calculate the number of sustained sprints made by each of the home team players. A sustained sprint is maintaining a speed >= 7 m/s for at least 1 second.

Here's our approach to the calculation part:

* We will consider a window of 25 frames and move it over the frame-wise speed values for a particular player. For a player to be considered making a sustained sprint, all of the speed values in the window have be at least equal to the threshold value of 7 m/s.


* For the window to move, we will add a new speed value to the window and remove the oldest speed value from the window. Every time we move our window over the speed values, we will check if all of the speed values in that window are at least equal to the threshold value or not, and keep track of this In the code below, we will use the *convolve* function from the *numpy* module to achieve this.


* Now, we have a record of the windows which qualify to be considered sustained sprints. Does the count of these windows give us the number of sustained sprints? No. For a player to make a sustained sprint, he needs to maintain a pace of 7 m/s for at least 1 second, but this sprint might extend up to several seconds. In such a case, that instance will be counted as a single sprint and not as multiple sprints of 1 second duration. 


* To calculate the number of sprints, we look for the instances where the player moved from a "no sustained sprint" window to a sustained sprint window. In other words, we have to count only those instances in which the player started his sprint and continued it for 1 second. The player might then continue his sprint in the next window(s) or not. If he does, it's still a part of the same sprint, and if he doesn't, it means that his sprint has ended. In the code below, we will use the *diff* and the *sum* functions of the *numpy* module to achieve this.

This is how we will count the number of sustained sprints for all of the home team players. Let's do this programmatically.

```python
num_sprints = []
sprint_threshold = 7 # threshold speed value (in m/s) for a sprint
sprint_window = 25 # window size; this refers to the minimum duration the sprint must be sustained, i.e. 25 frames or 1 second
for player in home_summary.index:
    column = 'Home_' + player + '_speed'
    player_sprints = np.diff(1*(np.convolve(1*(tracking_home[column] >= sprint_threshold), np.ones(sprint_window), mode='valid') == sprint_window))
    num_sprints.append(np.sum(player_sprints == 1))
home_summary['Number of Sustained Sprints'] = num_sprints
```

__4. Plot all the sustained sprints of a particular home team player.__

Player #8 has made the highest number of sustained sprints in the home team. Let's plot the trajectories of his 20 sprints.

```python
player = '8'
column = 'Home_' + player + '_speed'
column_x = 'Home_' + player + '_x'
column_y = 'Home_' + player + '_y'
player_sprints = np.diff(1*(np.convolve(1*(tracking_home[column] >= sprint_threshold), np.ones(sprint_window), mode='valid') == sprint_window))
player_sprints_start = np.where(player_sprints == 1)[0] + 1 # finding the frames where the sprints started
player_sprints_end = np.where(player_sprints == -1)[0] + sprint_window - 1 # finding the frames where the sprints ended

# creating the plot
fig,ax = mviz.plot_pitch()
for start, end in zip(player_sprints_start, player_sprints_end):
    ax.plot(tracking_home[column_x].iloc[start], tracking_home[column_y].iloc[start], 'ro')
    ax.plot(tracking_home[column_x].iloc[start:end+1], tracking_home[column_y].iloc[start:end+1], 'r')
```

__Note:__ The start of a sprint is shown by a red marker (or, filled-in circle).
